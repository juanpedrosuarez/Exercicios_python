Faça um programa para converter uma fração imprópria (representada por 2 números) em fração mista.

Entrada
A entrada contém dois números inteiros que representam o numerador e denominador de uma fração imprópria.

Saída
Seu programa deve imprimir na tela a conversão da fração imprópria para fração mista, conforme exemplo fornecido. Caso o denominador for igual a zero o programa deve imprimir “ERRO”, sem aspas.

Exemplos
Entrada

Saída

11

4

2(3/4)

10

3

3(1/3)

19

0

ERRO

```python
def converter(numerador:int, denominador:int):
    parte_inteira  = numerador // denominador
    novo_numerador = numerador %  denominador
    return (parte_inteira, novo_numerador, denominador)

def main():
    numerador   = int(input())
    denominador = int(input())
    if denominador!=0:
        (v1, v2, v3)   = converter(numerador, denominador)
        print(f"{v1}({v2}/{v3})")
    else:
        print("ERRO")

main()
```

```python
def fibonacci (num: int) -> bool:
   fib = 0
   fib1 = 0
   fib2 = 1

   while fib<num:
       fib = fib1 + fib2
       fib1 = fib2
       fib2 = fib

   return fib==num


num = int(input())

if fibonacci(num):
    print("Verdadeiro")
else:
    print("Falso")

```

```python
def conta_letras( frase: str):
    frase      = frase.lower()
    vogais     = 0
    espacos    = 0
    pontuacoes = 0
    consoantes = 0

    for i in frase:
        if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
            vogais += 1
        elif i == ' ':
            espacos += 1
        elif i == '.' or i == ',' or i == '!' or i == '?':
            pontuacoes += 1
        else:
            consoantes += 1

    comprimento = len(frase)

    print(f"Vogais: {vogais/comprimento*100:0.2f}%")
    print(f"Consoantes: {consoantes/comprimento*100:0.2f}%")
    print(f"Espacos: {espacos/comprimento*100:0.2f}%")
    print(f"Pontuacoes: {pontuacoes/comprimento*100:0.2f}%")


frase     = input()
conta_letras(frase)
```

```python
def leitura_da_matriz(n: int) -> list:
    M = list()
    for i in range(n):
        linha = list()
        for elemento in input().strip().split():
            linha.append(int(elemento))
        M.append(linha)
    return M

def minimo_na_coluna(matriz:list, col:int) -> int:
    menor = matriz[0][col]
    for i in range(len(matriz)):
        if matriz[i][col] < menor:
            menor = matriz[i][col]
    return menor

def soma_dos_menores(matriz:list):
    soma = 0
    for j in range(len(matriz[0])):
        soma += minimo_na_coluna(matriz, j)
    print(soma)

def main():
    n      = int(input())
    matriz = leitura_da_matriz(n)
    soma_dos_menores(matriz)

main()

```

```python
def leitura_da_matriz(n:int) -> list:
    M = [0]*n
    for i in range(n):
        L = input().split()
        for j in range(n):
            L[j] = int(L[j])
        M[i] = L
    return M

def somaX(M: list) -> int:
    n = len(M)
    soma = 0
    j = 0
    for i in range(n):
        soma = soma + M[i][j] + M[n-1-j][j]
        j = j + 1
    miolo = n//2
    soma = soma - M[miolo][miolo]
    return soma

def main():
    n       = int(input())
    M       = leitura_da_matriz(n)
    resposta = somaX(M)
    print(f"X = {resposta}")

main()
```

```python
def leitura_da_matriz(n: int) -> list:
    M = list()
    for i in range(n):
        linha = list()
        for elemento in input().strip().split():
            linha.append(int(elemento))
        M.append(linha)
    return M

def media_da_coluna(M: list, k: int) -> list:
    soma = 0.0
    for i in range(len(M)):
        soma += M[i][k]
    return soma/len(M)

def main():
    dimensao = input().split(" ")
    n = int(dimensao[0])
    m = int(dimensao[1])

    matriz = leitura_da_matriz(n)

    for j in range(m):
        quantidade = 0
        media = media_da_coluna(matriz, j)
        for i in range(n):
            if matriz[i][j]<media:
                quantidade += 1
        print(quantidade, end=" ")

main()
```
